```python
import pandas as pd
import numpy as np
import spacy
from  nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
import nltk
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
#load data
import pandas as pds
path='/content/Education.csv'
data=pd.read_csv(path)
data.head()
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
lematizer=WordNetLemmatizer()
lb=LabelEncoder()
cv=CountVectorizer(max_features=2000)
print(stopwords.words('english'))
```

```python
def preprocessingText(text):
    # Remove special characters
    text=re.sub("[^a-z A-Z 0-9]","",text)
    # Lower case convertion
    text=text.lower()
    return text
```

```python
def preprocessingTextUsingNLTK(text):
    text=preprocessingText(text)
    allEngStopWords=stopwords.words("english")
    #removal of stopwords and Lemmatizing
    if text!=None:
        text= " ".join([ lematizer.lemmatize(word) for word in text.split(" ") if word not in set(allEngStopWords) ])
    else:
        print(text)
    return text
```

```python
def preprocessingTextUsingSPACY(text):
    text=preprocessingText(text)
    lemWithStop=spacy.load("en_core_web_sm")
    doc=lemWithStop(text)
    #removal of stopwords and Lemmatizing
    if doc!=None and len(doc):
        text= " ".join([ word.lemma_ for word in doc if not word.is_stop ])
    else:
        print(text)
    return text
```

```python
data['Text']=data['Text'].apply(preprocessingTextUsingSPACY)
# df['Text']=df['Text'].apply(preprocessingTextUsingNLTK)
y=lb.fit_transform(data['Label'])
data.drop(columns=['Label'],inplace=True)
arr=cv.fit_transform(data['Text']).toarray()
X=pd.DataFrame(arr)
X.head()
```

```python
y=y.reshape(-1,1)
y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=20,test_size=0.2)
gnb=GaussianNB()
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
pred="I don't like studying math."
pred=preprocessingTextUsingSPACY(pred)
pred=cv.transform([pred]).toarray()
gnb.predict(pred)
if gnb.predict(pred)==1:
    print("Positive Education tweet")
else:
    print("Negative Education tweet")
```
